##### Q4. Write a program that prompts the user to enter a loan amount, annual interest rate, and number of years  for a car loan. Then calculate and print the monthly payment amount.

Note:
```
PMT= [PV*i*((1+i)^n)]/((1+i)^n)-1
```
PMT is the monthly payment

PV is the loan amount

i is the interest rate per month in decimal form (interest rate percentage divided by 12)

n is the number of months (term of the loan in months)

```
Loan Amount: 30000.00
Annual Interest Rate: 4.00
Years: 5
Your monthly payment is: $552.50
```

````
# Type solution here
PV = input("Loan Amount: ")

while PV.isalpha():                              # ensures input is a number
  print("You did not enter a number. Try again!")
  PV = input("Loan Amount: ")

i = input("Annual Interest Rate: ")

while i.isalpha():                               # ensures input is a number
  print("You did not enter a number. Try again!")
  i = input("Annual Interest Rate: ")

years = input("Years: ")

while years.isalpha():                           # ensures input is a number
  print("You did not enter a number. Try again!")
  years = input("Years: ")

PV = float(PV)
i = (float(i)/100)/12                            # converts interest rate into percentage and divides by 12
years = float(years)
n = years*12                                     # turns years into months

numerator = i*((1+i)**n)                         # solving equation (num)
denominator = ((1+i)**n)-1                       # solving equation (den)

finPMT = PV * (numerator/denominator)            # puts equation together

output = round(finPMT, 2)                        # rounds to two decimal places (like example)
           
print("Your monthly payment is: $", output)      # prints output
````

##### Q5. Write a program to that prompts the user to enter a string. Then check whether or not the string is a palindrome.

```
Enter a string: Madam
Madam, is a palindrome!
```

````
# Type solution here
userPhrase = input("Enter a string: ")            # takes a user input
revUserPhrase = userPhrase[::-1]                  #reverses user's input

if userPhrase.lower() == revUserPhrase.lower():   # checks if reversed word and actual word are equal, if they are it is a palindrome!
  print(userPhrase, "is a palindrome!")
else:
  print(userPhrase, "is not a palindrome!")
````

##### Q6. Write a program to compute and print all possible combinations of change for \$1. Denominations to be considered – quarter, dime, nickel, penny.

Hint: Use nested loops (loops within loops for the various denominations of coins)

```
Change for $1:
0 quarters, 0 dimes, 0 nickels, 100 pennies
…
4 quarters, 0 dimes, 0 nickels, 0 pennies
```

````
# Type solution here
print("Change for $1: ")
i = 0
j = 0
k = 0
l = 0
count = 0
for l in range(0, 5):                                                                   # loops through quarters
  l = l * 25
  for k in range(0,11):                                                                 # loops through dimes
    k = k * 10
    for j in range(0,21):                                                               # loops through nickles
      j = j * 5
      for i in range(0,101):                                                            # loops through pennies
        i = i * 1
        if i + j + k + l == 100:                                                        # ensures output is equal to 100, before printing
          print("Quarters", l//25, "Dimes", k//10, "Nickles", j//5, "Pennies", i//1)    # prints output
          count+= 1                                                                     # ensures the correct number of outputs is present
print(count)

````

##### Q8. Write a program to ask the user to enter a password. Then check to see if it is a valid password based on these requirements -

Must be at least 8 characters long

Must contain both uppercase and lowercase letters

Must contain at least one number between 0-9

Must contain a special character -!,@,#,$

If the password is not valid, ask the user to re-enter. This should continue until the user enters a valid password. After a valid password is entered, print Access Granted!

```
Please enter a password. Follow these requirements - 
a. Must be at least 8 characters long
b. Must contain both uppercase and lowercase letters
c. Must contain at least one number between 0-9
d. Must contain a special character -!,@,#,$
Password: HelloWorld1
Invalid password. Try again!
Password: Hello@World1
Access Granted!
```


````
# Type solution here
print("Please enter a password. Follow these requirements - \na. Must be at least 8 characters long \nb. Must contain both uppercase and lowercase letters \nc. Must contain at least one number between 0-9 \nd. Must contain a special character -!,@,#,$")
specialCharacter = "-!@#$"                                                              # these are which special characters the guest must use
bool = True
while bool:                                                                             # keeps asking user for new password if requirements aren't met
  userPass = input("Password: ")                                                        # takes user input
  if len(userPass) < 8:                                                                 # makes sure password is at least 8 letters
    print("Invalid Password. Your password is not at least 8 letters. Try again!")
  elif not any(i.islower() for i in userPass):                                          # makes sure there is a lower case letter in password
    print("Invalid Password. Your password has no lower case letter. Try again!")
  elif not any(i.isupper() for i in userPass):                                          # makes sure there is an upper case letter in password
    print("Invalid Password. Your password has no upper case letter. Try again!")
  elif not any(i.isdigit() for i in userPass):                                          # makes sure there is a number in the password
    print("Invalid Password. Your password has no numbers from 0-9. Try again!")  
  elif not any(i in specialCharacter for i in userPass):                                # makes sure there is one of the selected special characters in the password
    print("Invalid Password. Your password has no special characters -,!,@,#, or $. Try again!")  
  else:
    print("Password:", userPass, "\nAccess Granted!")                                   # grants access if all requirements are met
    bool = False

````
